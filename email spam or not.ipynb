```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Sample dataset
data = {
    "email": [
        "Win $1000 now!",
        "Your invoice is attached.",
        "You have been selected for a prize.",
        "Important: Update your account information.",
        "Meeting schedule for tomorrow.",
        "Claim your free vacation today!",
    ],
    "label": ["spam", "not spam", "spam", "spam", "not spam", "spam"],
}
```

```python
# Create a DataFrame
df = pd.DataFrame(data)

# Encode labels
df['label'] = df['label'].map({'spam': 1, 'not spam': 0})
```

```python
# Split data into features and labels
X = df['email']
y = df['label']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Convert text to numerical features
vectorizer = TfidfVectorizer()
X_train_transformed = vectorizer.fit_transform(X_train)
X_test_transformed = vectorizer.transform(X_test)
```

```python
# Train a Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train_transformed, y_train)

# Make predictions
y_pred = model.predict(X_test_transformed)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Test with a new email
new_email = ["Congratulations, you have won a free car!"]
new_email_transformed = vectorizer.transform(new_email)
prediction = model.predict(new_email_transformed)

print(f"The email is classified as: {'spam' if prediction[0] == 1 else 'not spam'}")
```
